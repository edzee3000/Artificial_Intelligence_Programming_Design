# example 1: anscombe

```python
anscombe = sns.load_dataset("anscombe")
# anscombe = pd.read_csv("seaborn-data/anscombe.csv")
```

```python
anscombe
```

```python
anscombe.info()
```

```python
anscombe.groupby("dataset").describe()
```

```python
anscombe.groupby("dataset").count()
```

```python
print(anscombe['dataset'] == 'I')
dataset_1 = anscombe[ anscombe['dataset'] == 'I']
dataset_1
```

```python
plt.plot(dataset_1['x'], dataset_1['y'], "bo")
```

```python
def plot_anscombe(dataset):
    plt.plot(dataset['x'], dataset['y'], 'o')
```

```python
dataset_2 = anscombe[ anscombe['dataset'] == 'II']
plot_anscombe(dataset_2)
```

```python
dataset_3 = anscombe[ anscombe['dataset'] == 'III']
plot_anscombe(dataset_3)
```

```python
dataset_4 = anscombe[ anscombe['dataset'] == 'IV']
plot_anscombe(dataset_4)
```

```python
fig = plt.figure()

subplot1 = fig.add_subplot(2, 2, 1)
subplot2 = fig.add_subplot(2, 2, 2)
subplot3 = fig.add_subplot(2, 2, 3)
subplot4 = fig.add_subplot(2, 2, 4)

def plot_anscombe(dataset, plt = plt):
    plt.plot(dataset['x'], dataset['y'], 'o')

plot_anscombe(dataset_1, subplot1)
plot_anscombe(dataset_2, subplot2)
plot_anscombe(dataset_3, subplot3)
plot_anscombe(dataset_4, subplot4)
```

```python
fig = plt.figure()

datasets = [dataset_1, dataset_2, dataset_3, dataset_4]

for i in range(1, 5):
    subp = fig.add_subplot(2, 2, i)
    plot_anscombe(datasets[i - 1], subp)
```

# example 2: flights

```python
#flights = sns.load_dataset("flights")

flights = pd.read_csv("seaborn-data/flights.csv")
print(flights)
flights = flights.loc[:,["year", "month", "passengers"]]

flights
```

```python
flights.info()
```

```python
flights.describe()
```

```python
flights["passengers"].plot()
```

## plot passengers by year

```python
flights.groupby("year").mean()
```

```python
flights.groupby("year").mean().plot()
```

```python
flights.groupby("year").std().plot()
```

```python
flights.groupby("month").mean()
```

## plot passengers by month

```python
pa = flights.groupby("month").mean()["passengers"]
pa
```

```python
pa.plot()
```

```python
pa.plot.bar()
```

### sorting data by month (1)

```python
flights[0:12].month
```

```python
flights.month
```

```python
monthdict = dict(zip(flights[0:12].month, range(1,13)))
monthdict
```

```python
pa_sorted = pa.rename(monthdict)
pa_sorted
```

```python
pa_sorted = pa_sorted.sort_index()
pa_sorted
```

```python
pa_sorted.plot.bar()
```

### ### sorting data by month (2)

```python
monthlist = [monthdict[m] for m in pa.index]
monthlist
```

```python
newpa = pd.DataFrame({"passenger":pa, "index":monthlist})
newpa
```

```python
newpa = newpa.sort_values("index")
newpa
```

```python
newpa["passenger"].plot.bar()
```

### plot with sns

for more details: https://seaborn.pydata.org/examples/timeseries_facets.html

```python
g = sns.relplot(
    data=flights,
    x="month", y="passengers", col="year", hue="year",
    kind="line",  linewidth=4, zorder=5,
    col_wrap=3, height=2, aspect=1.5, legend=False,
)

```
