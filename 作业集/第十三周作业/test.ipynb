```python
#Smith标准正交化
import sympy as sy
def inte(g,h):
    '''定义从-1到1上的定积分函数'''
    return sy.integrate(g*h,(x,-1,1))
x=sy.symbols("x")
x0,x1,x2,x3=1,x,x**2,x**3
y0=x0
y1=x1-inte(x1,y0)/inte(y0,y0)*y0
y2=x2-inte(x2,y1)/inte(y1,y1)*y1-inte(x2,y0)/inte(y0,y0)*y0
y3=x3-inte(x3,y2)/inte(y2,y2)*y2-inte(x3,y1)/inte(y1,y1)*y1-inte(x3,y0)/inte(y0,y0)*y0
print("经过Smith正交化后结果为：")
print("y0为：",y0,"y1为：",y1,"y2为：",y2,"y3为：",y3)
y0=y0/inte(y0,y0)
y1=y1/inte(y1,y1)
y2=y2/inte(y2,y2)
y3=y3/inte(y3,y3)
print("经过单位化之后结果为：")
print("y0为：",y0,"y1为：",y1,"y2为：",y2,"y3为：",y3)
```
